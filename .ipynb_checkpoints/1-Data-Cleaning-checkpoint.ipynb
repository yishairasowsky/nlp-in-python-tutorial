```python
# Web scraping, pickle imports
import requests
from bs4 import BeautifulSoup
import pickle
```

```python
def url_to_bible_transcript(url):
    '''Returns bible_transcript text specifically from mechon mamre'''
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    text = [elem.get_text() for elem in soup.find_all('p')]
    print(url)
    return text
```

```python
indices = [("%02d" % number) for number in range(1,36)]
```

```python
# URLs of transcripts in scope
bible_urls = [f"https://www.mechon-mamre.org/e/et/et{index}.htm" for index in indices]
# bible_urls
```

```python
import re
book_name_str = """Genesis, Exodus, Leviticus, Numbers, Deuteronomy
Joshua, Judges, Samuel, Kings, Isaiah, Jeremiah, Ezekiel, Hosea, Joel, Amos, Obadiah, Jonah, Micah, Nahum, Habakkuk, Zephaniah, Haggai, Zechariah, Malachi
Chronicles, Psalms, Job, Proverbs, Ruth, Song-of-Songs, Ecclesiastes, Lamentations, Esther, Daniel, Ezra-Nehemiah"""

book_name_str = re.sub(r'([^\s\w]|_)+', '', book_name_str )
bible_books = book_name_str.split()[0:]
# len(book_name_lst)
```

```python
# # Actually request transcripts (takes a few minutes to run)
bible_transcripts = [url_to_bible_transcript(u) for u in bible_urls[0:]]
```

## Cleaning The Data

```python
# Let's take a look at our data again
# next(iter(bible_data.keys()))
```

```python
# Notice that our dictionary is currently in key: comedian, value: list of text format
next(iter(bible_data.values()))
```

```python
# We are going to change this to key: comedian, value: string format
def combine_text(list_of_text):
    '''Takes a list of text and combines them into one large chunk of text.'''
#     combined_text = ' '.join(list_of_text)
    combined_text = ''.join(list_of_text)
    return combined_text
```

```python
# Combine it!
bible_data_combined = {key: [combine_text(value)] for (key, value) in bible_data.items()}
```

```python
# We can either keep it in dictionary format or put it into a pandas bible_dataframe
import pandas as pd
pd.set_option('max_colwidth',150)

bible_data_df = pd.DataFrame.from_dict(bible_data_combined).transpose()
bible_data_df.columns = ['transcript']
bible_data_df = bible_data_df.sort_index()
bible_data_df.sample(5)
```

```python
# Let's take a look at the transcript for Ali Wong
bible_data_df.transcript.loc['Genesis']
```

```python
# Apply a first round of text cleaning techniques
import re
import string

def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# Let's take a look at the updated text
bible_data_clean = pd.DataFrame(bible_data_df.transcript.apply(round1))
bible_data_clean
```

```python
# Apply a second round of cleaning
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
# Let's take a look at the updated text
bible_data_clean = pd.DataFrame(bible_data_clean.transcript.apply(round2))
bible_data_clean.sample(5)
```

```python
# # Let's take a look at our dataframe
bible_data_df.head()
```

```python
# Let's take a look at our dataframe
bible_data_df = bible_data_clean
bible_data_df.head()
```

```python
# bible_data_df.transcript.Amos
del bible_data_df['total_chars']
```

```python
# Let's add the number of words as well
full_names = bible_books
total_words = [len(bible_data_df.transcript[book].split()) for book in full_names]
bible_data_df['total_words'] = total_words
bible_data_df.head()
```

```python
# Let's pickle it for later use
bible_data_df.to_pickle("corpus.pkl")
```

```python
# We are going to create a document-term matrix using CountVectorizer, and exclude common English stop words
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
bible_data_cv = cv.fit_transform(bible_data_clean.transcript)
bible_data_dtm = pd.DataFrame(bible_data_cv.toarray(), columns=cv.get_feature_names())
bible_data_dtm.index = bible_data_clean.index
bible_data_dtm.sample(5)
```

```python
# Let's pickle it for later use
bible_data_dtm.to_pickle("dtm.pkl")
```
